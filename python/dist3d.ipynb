```python
import numpy as np

DX = 0.5

NX = 120

xmin = -4
xmax = 4
ymin = xmin
ymax = xmax

zmin = xmin
zmax = xmax

x1d = np.arange( xmin, xmax+DX, DX) 
y1d = np.arange( ymin, ymax+DX, DX)
z1d = np.arange( zmin, zmax+DX, DX)

x3d, y3d, z3d = np.meshgrid( x1d, y1d, z1d )

print(z3d.shape)
```

```python

dist3d = np.sqrt( np.square(x3d) + np.square(y3d ) + np.square(z3d) )


cz = 8
#cz = 20

import matplotlib.pyplot as plt

fig, ((ax1) ) = plt.subplots(1, 1, figsize=( 6,6) )

SCAT = ax1.scatter(x3d[:,:,cz], y3d[:,:,cz], s=0.5, color='k')
levs = [ 1, 2, 3, 4, 5]
levs = np.arange( 1, 11)
CONT = ax1.contour(x3d[:,:,cz], y3d[:,:,cz], dist3d[:,:,cz], levels=levs, colors='k',
                linewidths=1.0)

ax1.clabel(CONT, CONT.levels, fontsize=12, inline_spacing=0, 
          inline=True, fmt='%3.0fkm')
ax1.set_xlim( xmin, xmax )
ax1.set_ylim( ymin, ymax )

xlab = "X (km)"
ax1.set_xlabel(xlab, fontsize=12)
ylab = "Y (km)"
ax1.set_ylabel(ylab, fontsize=12)

ax1.hlines( y=0, xmin=xmin, xmax=xmax, ls='dashed', lw=1.0, colors='gray')
ax1.vlines( x=0, ymin=ymin, ymax=ymax, ls='dashed', lw=1.0, colors='gray')

# obs
odx = 0.5
odz = 0.5
ox1d = np.arange(xmin, xmax, odx)
oy1d = np.arange(ymin, ymax, odx)
oz1d = np.arange(zmin, zmax, odz)


ox3d, oy3d, oz3d = np.meshgrid( ox1d, oy1d, oz1d )

odist3d = np.sqrt( np.square(ox3d) + np.square(oy3d) + np.square(oz3d) )

dxx = 2.0
for dxx in np.arange(0.4, 3, 0.2): 
    print( "nobs: {0:0=4}, within 3D dist <= {1:.1f} km".format( len(odist3d[odist3d <= dxx ]), dxx ) )
```

```python
th = 2
tv = 2
odist3d_th = odist3d[::th,::th,::tv]
for dxx in np.arange(0.4, 3, 0.2): 
    print( "nobs: {0:0=4}, within 3D dist <= {1:.1f} km, thinning {2:0=2} (in H) and {3:0=2} (in V)".format( len(odist3d_th[odist3d_th <= dxx ]), dxx, th, tv ) )
```
